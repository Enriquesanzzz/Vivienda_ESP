```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv("housing_final.csv", sep=";")
df = df.drop(columns=["renta_es_proyeccion", "renta_neta_anual"])

df.head()
```

El objetivo será explicar de qué dependen los precios del metro cuadrado

```python
y = df["precio_compra_m2"]
X = df[["precio_alquiler_m2", "renta_mensual_neta", "tipo_interes_hipoteca"]]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
y_a = df["precio_alquiler_m2"]
X_a = df[["precio_compra_m2", "renta_mensual_neta", "tipo_interes_hipoteca"]]
X_a = sm.add_constant(X_a)
model = sm.OLS(y_a, X_a).fit()
print(model.summary())
```

En el análisis exploratorio se observa un nivel elevado de multicolinealidad entre algunas variables, especialmente entre los indicadores de renta y los precios de compra/alquiler. Esto no afecta negativamente al objetivo del proyecto, ya que el propósito de estos modelos es explicativo y descriptivo, centrado en entender relaciones generales y generar visualizaciones interpretables. La multicolinealidad solo sería problemática si el objetivo fuera estimar efectos causales exactos, algo que no se pretende en este trabajo. Además, los modelos predictivos posteriores no se ven perjudicados por ello, por lo que puede considerarse aceptable dentro del alcance del estudio.
